# **Get Encoding of Categorical Features**

```python
df.replace({'Fuel':{'Diesel':0,'Petrol':1,'CNG':2,'LPG':3,'Electric':4}},inplace=True)
```

```python
df.replace({'Seller_Type':{'Individual':0,'Dealer':1,'Trustmark Dealer':2}},inplace=True)
```

```python
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
df.replace({'Owner':{'First Owner':0,'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3,'Test Drive Car':4}},inplace=True)
```

```python
#X=pd.get_dummies(X,columns=['Fuel','Seller_Type','Transmission','Owner'],drop_first=True)
```

# **Define y (dependent or label or target variable) and X (independent or features or attribute variable)**

```python
y=df['Selling_Price']
```

```python
y.shape
```

```python
y
```

```python
X=df[['Year','KM_Driven','Fuel','Seller_Type','Transmission','Owner']]
```

or use drop function to define X

```python
#X=df.drop(['Brand','Model','Selling_Price'],axis=1)
```

```python
X.shape
```

# **Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Get Model Train**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

**Get Model Prediction**

```python
y_pred=lr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# **Get Model Evaluation**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

R-Square is very low signifies need for model improvement 

Encourage viewers to find the probable reasons for model poor performance

# **Get Visualization of Actual Vs Predicted Results**

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,color = 'BLACK')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
